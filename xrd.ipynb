```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
# Télécharge la structure du matériau étudié, SrHfN2  mp-9383
with MPRester("dFloIuOFLKpmmcc3") as m:
    structure = m.get_structure_by_material_id("mp-9383")
```

```python
# Rayonnement X incident de CuKa de longueur d onde 1.54060 A
Xray = XRDCalculator(1.54060)

# Analyse du Diffractogramme 
DiffPat = Xray.get_pattern(structure)
Theta = DiffPat.x
Intensity = DiffPat.y
print('\033[1m')
print("On étudie ici le diffractogramme d'un rayonnement X de CuKa de longueur d'onde 1.54060 Å sur le matériau SrHfN2")
print("________________________________________________________________________________________________________________")
print('')
# 2\Theta
print("Les angles 2\u03B8 [°] de diffraction : "+'\033[0m')
print('')
print(Theta)
print('------------------------------------------------')
# Intensité
print('\033[1m'+"Les intensités de diffraction (normalisées) : "+'\033[0m')
print('')
print(Intensity)
print('')
print("Les intensités ont été normalisées par rapport à l'intensité la plus importante.")
print("Le cinquième pic du diffractogramme ("+'\033[1m'+"angle 2\u03B8 [°] = %.2f" %Theta[5]+'\033[0m'+") prend donc la valeur de 100.")
print("Ensuite, toutes les autres intensités sont calculées à partir de cette valeur de référence.")
print("Celà est représenté sur le graphe ci-dessous, avec toutes les intensités normalisées fonction des angles 2\u03B8 [°] :")
Xray.show_plot(structure)
```

```python
# Analyse des 3 premiers pics du diffractogramme
print('\033[1m'+"Nous nous intéressons maintenant plus précisément aux 3 premiers pics du diffractogramme ")
print("_________________________________________________________________________________________")
print('')
print('\033[1m'+"Le premier pic du diffractogramme : "+'\033[0m')
print('')
print("Un angle 2\u03B8 [°] : "+'\033[1m', Theta[0])
print('\033[0m')
print("Une intensité : "+'\033[1m', Intensity[0])
print('\033[0m')
print("Les indices de Miller (h,k,l) associés : " +'\033[1m', DiffPat.hkls[0][0]['hkl'])
print('\033[0m')
print('')
print("--------------------------------------")
print('\033[1m'+"Le second pic du diffractogramme : "+'\033[0m')
print('')
print("Un angle 2\u03B8 [°] : "+'\033[1m', Theta[1])
print('\033[0m')
print("Une intensité : "+'\033[1m', Intensity[1])
print('\033[0m')
print("Les indices de Miller (h,k,l) associés : " +'\033[1m', DiffPat.hkls[1][0]['hkl'])
print('\033[0m')
print('')
print("--------------------------------------")
print('\033[1m'+"Le troisième pic du diffractogramme : "+'\033[0m')
print('')
print("Un angle 2\u03B8 [°] : "+'\033[1m', Theta[2])
print('\033[0m')
print("Une intensité : "+'\033[1m', Intensity[2])
print('\033[0m')
print("Les indices de Miller (h,k,l) associés : " +'\033[1m', DiffPat.hkls[2][0]['hkl'])
print('\033[0m')
print('')
print("--------------------------------------")
print('')
DiffPat3 = Xray.get_pattern(structure,True,(0,32))
Intensity3 = DiffPat3.y
print("Les intensités calculées ci-dessus ont pour référence le cinquième pic du diffractogramme.")
print("On pourrait également les calculer en prenant comme référence le second pic, le plus intense des trois premiers.")
print("Cela nous donnerait pour les trois premiers pics respectivement les intensités suivantes : " +'\033[1m' + "%.4f, %.4f, %.4f." %(Intensity3[0],Intensity3[1],Intensity3[2]))
print('\033[0m'+"Ce graphe ci-dessous représente les intensités des 3 premiers pics du diffractogramme en fonction des angles 2\u03B8 [°], ")
print("Où les intensités sont normalisées avec comme intensité de référence le second pic :")
Xray.show_plot(structure,two_theta_range=(0,32))
```
